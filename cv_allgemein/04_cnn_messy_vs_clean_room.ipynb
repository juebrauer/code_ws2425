# CNN mit Keras zur Klassifikation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, InputLayer
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt
import numpy as np

# Pfad zum Datensatzordner
train_dataset_folder = "messy_vs_clean_room_dataset/train/"
val_dataset_folder = "messy_vs_clean_room_dataset/val/"

# Parameter
img_height, img_width = 150, 150
batch_size = 32
epochs = 50

# Trainingsdaten und Validierungsdaten mit Data Augmentation laden
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,       # Bilder zufällig um bis zu 20 Grad drehen
    width_shift_range=0.2,   # Bilder horizontal zufällig um bis zu 20% verschieben
    height_shift_range=0.2,  # Bilder vertikal zufällig um bis zu 20% verschieben
    shear_range=0.2,         # Scherung der Bilder zufällig um bis zu 20%
    zoom_range=0.2,          # Bilder zufällig um bis zu 20% zoomen
    horizontal_flip=True,    # Zufälliges horizontales Spiegeln
    fill_mode='nearest'      # Methode zur Füllung leerer Bildbereiche nach Transformationen
)

val_datagen = ImageDataGenerator(rescale=1.0/255)  # Nur Reskalierung für Validierungsdaten

# Trainings- und Validierungs-Generatoren
train_generator = train_datagen.flow_from_directory(
    train_dataset_folder,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

validation_generator = val_datagen.flow_from_directory(
    val_dataset_folder,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

show_different_augmented_images = False
plt.figure(figsize=(8, 8))
if show_different_augmented_images:    
    for i in range(16):        
        augmented_image, _ = next(train_generator)
        ax = plt.subplot(4, 4, i + 1)
        plt.imshow(augmented_image[0])
        plt.axis("off")    
else:
    sample_image, _ = next(train_generator)
    sample_image = sample_image[0]  # Nur das erste Bild im Batch extrahieren    
    for i in range(16):
        augmented_image = train_datagen.random_transform(sample_image)
        ax = plt.subplot(4, 4, i + 1)
        plt.imshow(augmented_image)
        plt.axis("off")
plt.tight_layout()
plt.show()


# CNN-Modell definieren
model = Sequential([
    InputLayer( shape=(img_height, img_width, 3) ),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Modell kompilieren
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# Modell trainieren
history = model.fit(
    train_generator,
    validation_data=validation_generator,    
    epochs=epochs
)

# Modellbewertung

# Validierungs-Data Augmentation Generator (nur zum Testen)
augmented_val_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Generator für augmentierte Validierungsbilder
augmented_validation_generator = augmented_val_datagen.flow_from_directory(
    val_dataset_folder,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

num_augmented_images = 500
augmented_images = []
augmented_labels = []

for i in range(num_augmented_images):
    if i % 100 == 0:
        print(f"generated {i} augmented validation images so far...")
    augmented_image, label = next(augmented_validation_generator)        
    augmented_images.append(augmented_image[0])  # [0] ist notwendig, da der Generator im Batch-Format zurückgibt
    augmented_labels.append(label[0])
augmented_images = np.array(augmented_images)
augmented_labels = np.array(augmented_labels)

augmented_loss, augmented_accuracy = model.evaluate(augmented_images, augmented_labels, batch_size=32)
print(f"Augmented Validation accuracy on 1000 images: {augmented_accuracy:.2f}")

# Modell speichern
model.save("messy_vs_clean_room_classifier.keras")
```
